```python
L = ["Ram",1,"Shyam",2,"Aman",3]
print(L)
L.sort()
print(L)
```

```python
# the problem is that the L.sort() tries to compare the strings and inetgers directly which python does'nt allow thats why it gives an error
# to fix this issue we will give sort() a key which will convert everything to a common, comparable form

L = ["Ram",1,"Shyam",2,"Aman",3]
print(L)
L.sort(key=str)
print(L)
```
